4. Data Cleaning

```python
# Checking for missing values
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percent (%)': missing_percent})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values(by='Percent (%)', ascending=False)
print(missing_df)
```

```python
# Define essential columns to never drop
essential_cols = ['country', 'year', 'co2', 'gdp']

# Clean up column names (e.g., whitespace issues)
df.columns = df.columns.str.strip()

# Drop sparse columns, excluding essential ones
sparse_cols = df.isnull().mean()
sparse_cols = sparse_cols[sparse_cols > 0.80].index.tolist()
sparse_cols = [col for col in sparse_cols if col not in essential_cols]

# Now drop
df_cleaned = df.drop(columns=sparse_cols)
```

```python
# Remove duplicates
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Remove non-country/listings with no valid iso-code
non_countries = [
    'World', 'Asia', 'Africa', 'Europe', 'European Union', 'North America',
    'South America', 'Oceania', 'International transport', 'High-income countries',
    'Upper-middle-income countries', 'Lower-middle-income countries', 'Low-income countries'
]

df_cleaned = df_cleaned[~df_cleaned['country'].isin(non_countries)]
df_cleaned = df_cleaned[df_cleaned['iso_code'].apply(lambda x: isinstance(x, str) and len(x) == 3)]
```

```python
# Keep only rows from 1990 onwards
df_cleaned = df_cleaned[df_cleaned['year'] >= 1990]
print(df_cleaned.head())
```

```python
# Drop countries with fewer than 20 years of CO2 data
valid_countries = df_cleaned.groupby('country')['co2'].count()
valid_countries = valid_countries[valid_countries >= 20].index
df_cleaned = df_cleaned[df_cleaned['country'].isin(valid_countries)]
print(df_cleaned.head())
```

```python
# Sort first
df_cleaned = df_cleaned.sort_values(['country', 'year'])

# Use transform instead of groupby+ffill to avoid losing columns
df_cleaned = df_cleaned.groupby('country').transform(lambda x: x.ffill())

# Re-attach 'country' and 'year' from the original df
df_cleaned['country'] = df['country']
df_cleaned['year'] = df['year']
```

```python
# Create derived features
df_cleaned['co2_per_gdp'] = df_cleaned['co2'] / df_cleaned['gdp']
df_cleaned['co2_change'] = df_cleaned.groupby('country')['co2'].pct_change()
```

```python
# Final cleaning checks
df_cleaned = df_cleaned.drop_duplicates(subset=['country', 'year'])
df_cleaned['year'] = df_cleaned['year'].astype(int)
df_cleaned = df_cleaned.set_index('country', drop=False)

# Move 'country' and 'iso_code' to the front
cols = ['country', 'iso_code'] + [col for col in df_cleaned.columns if col not in ['country', 'iso_code']]
df_cleaned = df_cleaned[cols]
```
