2. Univariate Analysis

#### 2.1 Numerical Variables

```python
# Summary statistics
df.describe()
```

```python
# Step 1: Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Step 2: Replace inf values with NaN
numeric_df = numeric_df.replace([np.inf, -np.inf], np.nan)

# Step 3: Drop columns that are all NaNs or constant
numeric_df = numeric_df.dropna(axis=1, how='all')  # drop all-NaN columns
numeric_df = numeric_df.loc[:, numeric_df.nunique(dropna=True) > 1]  # drop constant columns

# Step 4: Plot in batches
def plot_histograms(df, cols, batch_size=6):
    for i in range(0, len(cols), batch_size):
        subset = cols[i:i+batch_size]
        df[subset].hist(figsize=(18, 10), bins=30, edgecolor='black')
        plt.suptitle(f"Histograms: {', '.join(subset)}", fontsize=16)
        plt.tight_layout()
        plt.show()

# Step 5: Run it on cleaned numeric data
plot_histograms(numeric_df, numeric_df.columns.tolist())
```

Histograms show the frequency distribution (how values are spread out) and for time series modeling, we want to know information such as is GDP right-skewed? Is CO₂ growth mostly flat or heavy-tailed? Are any features mostly 0s or constants?

The reason we choose all numerics is that we don't yet know which variables are informative — so we start wide. Histograms help eliminate irrelevant features quickly.

#### 2.2 Categorical Variables

```python
# Count of countries
print("Number of countries:", df['country'].nunique())
df['country'].value_counts().head(10).plot(kind='barh')
plt.title("Top 10 Countries by Data Count")
plt.show()
```

We want to confirm that you have broad country representation and spot imbalance (some countries may have far more years of data than others)

Choosing the top 10 country bar plot helps visualize dominant entities in our dataset

3. Bivariate Analysis

#### 3.1 Correlation Matrix

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', center=0, annot=False)
plt.title("Correlation Matrix (Numeric Features)")
plt.show()
```

A heatmap gives us a fast overview of linear relationships and highlights potential multicollinearity (e.g. GDP vs energy vs population)

Helps with feature selection by eliminating strongly correlated pairs or engineering new composite features

#### 3.2 Scatter Plots

```python
# CO2 vs GDP
sns.scatterplot(data=df, x='gdp', y='co2', hue='country', legend=False)
plt.title("CO2 Emissions vs GDP")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

If gdp and co2 show a log-linear relationship, that’s critical for regression forecasting.

```python
# CO2 per capita vs energy per capita
sns.scatterplot(data=df, x='energy_per_capita', y='co2_per_capita', hue='country', legend=False)
plt.title("CO2 per Capita vs Energy per Capita")
plt.show()
```

co2_per_capita vs energy_per_capita gives a per-person impact view, helpful for policy implications.

### 4. Time Series Exploration

```python
# Example country time series
top_country = df['country'].value_counts().idxmax()
df_country = df[df['country'] == top_country]

plt.plot(df_country['year'], df_country['co2'], label='CO2 Emissions')
plt.plot(df_country['year'], df_country['gdp'], label='GDP')
plt.title(f"Time Series for {top_country}")
plt.xlabel("Year")
plt.legend()
plt.show()
```

You’re forecasting emissions to understand temporal patterns

Plotting co2 and gdp together helps test if GDP growth explains emission trends

We chose a country with the most data (value_counts().idxmax()) to ensure we have a rich time series

### 5. Feature Selection Ideas
